```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import seaborn as sns

```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)



neglim = mu - 4*sigma
poslim = mu + 4*sigma
x = np.linspace(neglim, poslim , 500)

plt.figure(figsize=(7,3), dpi=400)
#ax = plt.subplot(111)



# Hide the right and top spines
#ax.spines['right'].set_visible(True)
#ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
#ax.yaxis.set_ticks_position('left')
#ax.xaxis.set_ticks_position('bottom')

figure = plt.figure(figsize=(7,3.5), dpi =400)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='black')

section = np.linspace(mu -4*sigma, mu + 1.5*sigma, 500)
#section2 = np.linspace(mu + 0.2*sigma, poslim, 500)

plt.fill_between(section, stats.norm.pdf(section, mu, sigma), color='lightgrey', label="0.4207")
#plt.fill_between(section2, stats.norm.pdf(section2, mu, sigma), color='b', label="0.20")
#plt.legend(loc='upper left', frameon=False)

plt.xlabel('t')
plt.ylabel('f(t)',labelpad=10)
sns.despine() #removes upper and left axis

plt.text(-0.4,0.13, r' 1 - $\alpha$', size="larger")
plt.text(1.72,0.022, r'$\alpha$', size="larger")

#vertical black line at x=1.5
x2, y2 =  [1.5, 1.5], [0.000, stats.norm.pdf(1.5, mu, sigma)-0.001]
plt.plot( x2, y2, c= "black", linewidth = 1)

plt.show()
figure.savefig("t-verteilung.pdf",format='pdf')
```

```python
#PLOT TWO NORMAL DISTRIBUTIONS


mu_1 = 60
variance_1 = 4/16
sigma_1 = math.sqrt(variance_1)

mu_2 = 59
variance_2 =4/16
sigma_2 = math.sqrt(variance_2)


neglim = min(mu_1 - 4*sigma_1, mu_2 -4*sigma_2)
poslim = max(mu_1 + 4*sigma_1, mu_2 + 4*sigma_2)
x = np.linspace(neglim, poslim , 500)

plt.figure(figsize=(7,3), dpi=400)
#ax = plt.subplot(111)

#create figure:
figure = plt.figure(figsize=(7,3.5), dpi =400)

#plot curve 1:
plt.plot(x, stats.norm.pdf(x, mu_1, sigma_1), color='black')

#plot curve 2:
plt.plot(x, stats.norm.pdf(x, mu_2, sigma_2), color='black')

section = np.linspace(58.835, mu_1 + 4*sigma_1, 500)
#section2 = np.linspace(mu + 0.2*sigma, poslim, 500)

plt.fill_between(section, stats.norm.pdf(section, mu_2, sigma_2), color='orange', alpha = 0.5, label="0.4207")
#plt.fill_between(section2, stats.norm.pdf(section2, mu, sigma), color='b', label="0.20")
#plt.legend(loc='upper left', frameon=False)

#plt.xlabel(r'$\bar{x}$')
#plt.ylabel(r'f($\bar{x}$)',labelpad=10)

plt.xlabel(' ')
plt.ylabel(' ',labelpad=10)

sns.despine() #removes upper and left axis

#plt.text(-0.4,0.13, r' 1 - $\alpha$', size="larger")
#plt.text(1.72,0.022, r'$\alpha$', size="larger")

#vertical black line at x=60
#x2, y2 =  [60, 60], [0, stats.norm.pdf(60, mu_1, sigma_1)-0.001]
#plt.plot( x2, y2, c= "black", linewidth = 1)

#set ticks:
plt.xticks([58.835,60])
plt.yticks([])

plt.show()
figure.savefig("zwei_normalverteilungen_yellow2.pdf",format='pdf')
```
